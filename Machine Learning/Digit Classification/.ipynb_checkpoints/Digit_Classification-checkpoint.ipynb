Load the data. Notice that the data gets partitioned into training, development, and test sets. Also, a small subset of the training data called mini_train_data and mini_train_labels gets defined, which you should use in all the experiments below, unless otherwise noted.

```python
# Load the digit data from https://www.openml.org/d/554 or from default local location '~/scikit_learn_data/...'
X, Y = fetch_openml(name='mnist_784', return_X_y=True, cache=False)


# Rescale grayscale values to [0,1].
X = X / 255.0

# Shuffle the input: create a random permutation of the integers between 0 and the number of data points and apply this
# permutation to X and Y.
# NOTE: Each time you run this cell, you'll re-shuffle the data, resulting in a different ordering.
shuffle = np.random.permutation(np.arange(X.shape[0]))

X, Y = X[shuffle], Y[shuffle]

print('data shape: ', X.shape)
print('label swhape:', Y.shape)

# Set some variables to hold test, dev, and training data.
test_data, test_labels = X[61000:], Y[61000:]
dev_data, dev_labels = X[60000:61000], Y[60000:61000]
train_data, train_labels = X[:60000], Y[:60000]
mini_train_data, mini_train_labels = X[:1000], Y[:1000]
```

### Part 1:

Show a 10x10 grid that visualizes 10 examples of each digit.

Notes:
* You can use `plt.rc()` for setting the colormap, for example to black and white.
* You can use `plt.subplot()` for creating subplots.
* You can use `plt.imshow()` for rendering a matrix.
* You can use `np.array.reshape()` for reshaping a 1D feature vector into a 2D matrix (for rendering).

```python
def P1(num_examples=10):

### STUDENT START ###

    # create square grid of images with side length of num_examples
    fig, axes = plt.subplots(num_examples, num_examples, figsize = (10,10))
    
    # gets unique labels (i.e. digit names) from the data
    unique_labels = np.unique(test_labels)
    
    # iterator for tracking how many digits have been completed
    # note: int value not related to digit value
    # (i.e. current_digit = 0 does not start the grid with the digit zero)
    current_digit = 0
    
    # iterates over each of the ten digits
    for label in unique_labels:
        
        # gets indices in the data for current digit
        indices = np.where(test_labels == label)
        
        # iterates over each example requested for the grid
        for i in range(num_examples):

            # gets current image from test data at index
            # of the appropriate digit
            image = test_data[indices[0][i]]
            
            # reshapes data from 1D into a 28x28 image
            new_image = np.reshape(image, (28,28))
            
            # sets colormap to black and white
            plt.rc("image", cmap = "gray")
            
            # displays image on row with similar digits
            axes[current_digit, i].imshow(new_image)
            
            # turns off grid for subplot
            axes[current_digit, i].grid(False)
            
            # turns off x and y tick markers for subplot
            axes[current_digit, i].set_xticks([])
            axes[current_digit, i].set_yticks([])
        
        # iterates counter of completed digits
        current_digit += 1
    
    plt.show()
### STUDENT END ###

P1(10)
```

### Part 2:

Produce k-Nearest Neighbors models with k $\in$ [1,3,5,7,9].  Evaluate and show the accuracy of each model. For the 1-Nearest Neighbor model, additionally show the precision, recall, and F1 for each label. Which digit is the most difficult for the 1-Nearest Neighbor model to recognize?

Notes:
* Train on the mini train set.
* Evaluate performance on the dev set.
* You can use `KNeighborsClassifier` to produce a k-nearest neighbor model.
* You can use `classification_report` to get precision, recall, and F1 results.

```python
def P2(k_values):

### STUDENT START ###

    # list of dicts, initially empty, to contain each
    # k-classifier's report
    report_list = []

    # loops over each value of k
    for k in k_values:

        # create instance of k nearest neighbors classifier
        classifier = KNeighborsClassifier(k)
        
        # generate model from mini training data
        classifier.fit(mini_train_data, mini_train_labels)
        
        # generate predictions for dev images
        predictions = classifier.predict(dev_data)
        
        # generates report for the model
        report = sklearn.metrics.classification_report(dev_labels, predictions, output_dict = True)

        # overall accuracy for the model
        accuracy = np.round(100*report["accuracy"], 1)
        
        # prints accuracy found in the current model
        print("Model " + str(k) + " accuracy: " + str(accuracy))
        
        # appends report to the report list
        report_list.append(report)
        
        # extra reporting for the k = 1 model
        if k == 1:
            
            # get and iterate over each digit label
            for digit in [str(i) for i in range(10)]:
                print("Digit " + str(digit) + ":")
                
                # gets precision for the current digit
                precision = np.round(100*report[digit]["precision"], 1)
                
                # prints precision for the digit
                print("Precision = " + str(precision), end = ", ")
                
                # gets recall for the current digit
                recall = np.round(100*report[digit]["recall"], 1)
                
                # prints recall for the digit
                print("Recall = " + str(recall), end = ", ")
                
                # gets F1 score for the current digit
                f1 = np.round(100*report[digit]["f1-score"], 1)
                
                # prints F1 score for the digit
                print("F1 score = " + str(f1))
        
### STUDENT END ###

k_values = [1, 3, 5, 7, 9]
P2(k_values)
```

ANSWER: The digit 8 / Eight was the most difficult for the k = 1 nearest neighbor algorithm to detect as it had the lowest F1 score out of all the digits, a statistic which combines the precision and recall for each.

### Part 3:

Produce 1-Nearest Neighbor models using training data of various sizes.  Evaluate and show the performance of each model.  Additionally, show the time needed to measure the performance of each model.

Notes:

* Evaluate on the dev set.
* You can use `KNeighborsClassifier` to produce a k-nearest neighbor model.
* You can use `time.time()` to measure elapsed time of operations.

```python
def P3(train_sizes, accuracies):

### STUDENT START ###

    # k value for using first nearest neighbor model
    k = 1 
    
    # create instance of k nearest neighbors classifier
    classifier = KNeighborsClassifier(k)
    
    # iterate over each training size
    for size in train_sizes:
        
        time_one = time.time()
        
        # subset the training data
        training_data_subset = train_data[:size]
        
        # subset the training labels
        training_labels_subset = train_labels[:size]
        
        # generate model from subset of training data
        classifier.fit(training_data_subset, training_labels_subset)
        
        # generate predictions for dev images
        predictions = classifier.predict(dev_data)
        
        # generates report for the model
        report = sklearn.metrics.classification_report(dev_labels, predictions, output_dict = True)
        
        # overall accuracy for the model
        accuracy = np.round(100*report["accuracy"], 1)
        
        time_two = time.time()
        
        net_time = time_two - time_one
        
        # prints accuracy found in the current model
        print("Training data size: " + str(size) + ", Accuracy: " + str(accuracy) +
             ", Time (s): " + str(net_time))
        

### STUDENT END ###

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25600]
accuracies = []
P3(train_sizes, accuracies)
```

### Part 4:

Produce a linear regression model that predicts accuracy of a 1-Nearest Neighbor model given training set size. Show $R^2$ of the linear regression model.  Show the accuracies predicted for training set sizes 60000, 120000, and 1000000.  Show a lineplot of actual accuracies and predicted accuracies vs. training set size over the range of training set sizes in the training data.  What's wrong with using linear regression here?

Apply a transformation to the predictor features and a transformation to the outcome that make the predictions more reasonable.  Show $R^2$ of the improved linear regression model.  Show the accuracies predicted for training set sizes 60000, 120000, and 1000000.  Show a lineplot of actual accuracies and predicted accuracies vs. training set size over the range of training set sizes in the training data - be sure to display accuracies and training set sizes in appropriate units.

Notes:
* Train the linear regression models on all of the (transformed) accuracies estimated in Problem 3.
* Evaluate the linear regression models on all of the (transformed) accuracies estimated in Problem 3.
* You can use `LinearRegression` to produce a linear regression model.
* Remember that the sklearn `fit()` functions take an input matrix X and output vector Y. So, each input example in X is a vector, even if it contains only a single value.
* Hint re: predictor feature transform: Accuracy increases with training set size logarithmically.
* Hint re: outcome transform: When y is a number in range 0 to 1, then odds(y)=y/(1-y) is a number in range 0 to infinity.

```python
def P4():

### STUDENT START ###
    
    
    ### This section creates the first linear model ###
    
    # accuracies found from Part 3 above
    accuracies = np.array([70.2, 79.1, 81.1, 86.6, 90.5, 92.7, 93.9, 95.2, 96.3])
    
    # reshapes input into 2D array as required by the linear regression function
    train_sizes_reshaped = np.reshape(train_sizes, (-1, 1))

    # fits linear model with training sizes and independent variable
    # and the accuracies as the dependent variable
    regression = LinearRegression().fit(train_sizes_reshaped, accuracies)
    
    # gets r-squared value of the regression and rounds to two decimals
    r_squared = np.round(regression.score(train_sizes_reshaped, accuracies),2)
    
    print("The R-Squared value of the first linear regression is " + str(r_squared))
    
    # new training set sizes to predict
    new_sizes = np.array([[60000], [120000], [1000000]])
    
    # predictions of the new training set sizes
    predictions = regression.predict(new_sizes)
    
    print("The predicted accuracies for the larger training sizes are " + 
          str(np.round(predictions[0],0)) + ", " + 
          str(np.round(predictions[1],0)) + ", and " + 
          str(np.round(predictions[2],0)) + 
          " which are not possible as they are greater than 100%.")
    
    
    
    
    ### This section creates the second linear model ###
    
    # takes the base 10 log of the input
    transformed_input = np.log10(train_sizes_reshaped)
    
    # takes the odds function of the decimal-format accuracies
    transformed_output = (accuracies/100.)/(1 - (accuracies/100.))

    # fits linear model with transformed input/output
    transformed_regression = LinearRegression().fit(transformed_input, transformed_output)
    
    # gets predicted values of the regression model with the odds function still applied
    odds_predictions = transformed_regression.predict(transformed_input)
    
    # takes inverse of the accuracy transformation to get the true predicted accuracies
    nonodds_predictions = 100*(odds_predictions/(1 + odds_predictions))
    
    # gets r-squared value of the regression and rounds to two decimals
    transformed_r_squared = np.round(transformed_regression.score(transformed_input, transformed_output),2)
    
    print("The R-Squared value of the transformed linear regression is " + str(transformed_r_squared))
    
    # gets the transformation of the new input sizes
    transformed_new_sizes = np.log10(new_sizes)
    
    # predictions of the new transformed input sizes
    transformed_predictions = transformed_regression.predict(transformed_new_sizes)
    
    # undo accuracy transformation on new predictions
    new_nonodds_predictions = 100*(transformed_predictions/(1 + transformed_predictions))
    
    print("The predicted accuracies for the larger training sizes are " + 
          str(np.round(new_nonodds_predictions[0],0)) + ", " + 
          str(np.round(new_nonodds_predictions[1],0)) + ", and " + 
          str(np.round(new_nonodds_predictions[2],0)) + 
          " which are all possible as they are not over 100%.")
        
        

    ### This section creates the plots ###
    
    # creates fig and axes for the two plots
    fig, axes = plt.subplots(2, 1, sharex = True, sharey = True, figsize = (8,8))
    
    # adds true accuracies to the first plot
    axes[0].plot(train_sizes, accuracies, "ro")
    
    # adds predicted accuracies to the first plot
    axes[0].plot(train_sizes, regression.predict(train_sizes_reshaped), "-b")
    
    # adds true accuracies to the second plot
    axes[1].plot(train_sizes, accuracies, "ro")
    
    # adds predicted accuracies to the second plot
    axes[1].plot(train_sizes, nonodds_predictions, "-b")
    
    # limits second y axis
    axes[1].set_ylim([0, 100])
    
    # add a big axis for common x and y titles
    fig.add_subplot(111, frameon=False)
    plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
    plt.xlabel("Training Data Set Size", fontsize = 15)
    plt.ylabel("Accuracy (%)", fontsize = 15)
    plt.title("Accuracy versus Training Data in K = 1 NN Model", fontsize = 15)
    
    plt.show()

### STUDENT END ###

P4()
```

ANSWER: The mistake with using the linear regression on the non-transformed data is that the line will predict that increases in training size will continually increase accuracy beyond 100%. However, the maximum possible accuracy by definition is 100%. Transforming the input to scale logarithmically with the output makes all accuracies within the range of interest still beneath 100 and better fits the data.

### Part 6:

A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian, i.e., the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur filter by just using the 8 neighboring pixels like this: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values.

Pick a weight, then produce and evaluate four 1-Nearest Neighbor models by applying your blur filter in these ways:
- Do not use the filter
- Filter the training data but not the dev data
- Filter the dev data but not the training data
- Filter both training data and dev data

Show the accuracies of the four models evaluated as described.  Try to pick a weight that makes one model's accuracy at least 0.9.

Notes:
* Train on the (filtered) mini train set.
* Evaluate performance on the (filtered) dev set.
* There are other Guassian blur filters available, for example in `scipy.ndimage.filters`. You are welcome to experiment with those, but you are likely to get the best results with the simplified version described above.

```python
def P6():
    
### STUDENT START ###

    # weight for Gaussian blurring
    weight = 0.8
    
    ### First Model ### 
    
    # sets value of k for NN model
    k = 1

    # create instance of k nearest neighbors classifier
    classifier_one = KNeighborsClassifier(k)

    # generate model from mini training data
    classifier_one.fit(mini_train_data, mini_train_labels)

    # generate predictions for dev images
    predictions_one = classifier_one.predict(dev_data)
        
    # generates report for the model
    report_one = sklearn.metrics.classification_report(dev_labels, predictions_one, output_dict = True)

    # overall accuracy for the model
    accuracy_one = np.round(100*report_one["accuracy"], 1)
    
    print("Model one (no filter) accuracy: " + str(accuracy_one))
    
    
    ### Second Model ###
    
    # create empty array to hold filtered training data
    filtered_train_data = np.empty_like(mini_train_data)
    
    # counter to keep track of image location in the dataset
    image_index = 0
    
    # loops over each image to apply the filter
    for image in mini_train_data:
        
        # counter to keep track of pixel location in image
        pixel_index = 0
        
        # create blank new image
        new_image = np.empty_like(image)
    
        # loops over each pixel 
        for pixel in image:
            
            if pixel_index < 754: # for each row except the last
                
                # takes average of the surrounding pixels
                surrounding_pixels_value = np.average(image[pixel_index+1] + image[pixel_index-1] +
                                                      image[pixel_index+28] + image[pixel_index+27] +
                                                      image[pixel_index+29] +
                                                      image[pixel_index-28] + image[pixel_index-27] +
                                                      image[pixel_index-29])
            else: # for the last row
                
                # takes average of the surrounding pixels
                surrounding_pixels_value = np.average(image[pixel_index-1] +
                                                      image[pixel_index-28] + image[pixel_index-27] +
                                                      image[pixel_index-29])
            
            # determines new pixel weight by linear conbination with coefficient weight
            new_pixel = weight*pixel + (1-weight)*surrounding_pixels_value
            
            # replaces old pixel with the new pixel value
            new_image[pixel_index] = new_pixel
            
            # increments pixel location in the image
            pixel_index += 1
            
        # append new image
        filtered_train_data[image_index] = new_image
        
        # increments the counter
        image_index += 1

    # create instance of k nearest neighbors classifier
    classifier_two = KNeighborsClassifier(k)

    # generate model from filtered training data
    classifier_two.fit(filtered_train_data, mini_train_labels)

    # generate predictions for dev images
    predictions_two = classifier_two.predict(dev_data)
        
    # generates report for the model
    report_two = sklearn.metrics.classification_report(dev_labels, predictions_two, output_dict = True)

    # overall accuracy for the model
    accuracy_two = np.round(100*report_two["accuracy"], 1)
    
    print("Model two (training data filtered) accuracy: " + str(accuracy_two))
    
    
    
    ### Third Model ###
    
    # create empty array to hold filtered dev data
    filtered_dev_data = np.empty_like(dev_data)
    
    # counter to keep track of image location in the dataset
    image_index = 0
    
    # loops over each image to apply the filter
    for image in dev_data:
        
        # counter to keep track of pixel location in image
        pixel_index = 0
        
        # create blank new image
        new_image = np.empty_like(image)
    
        # loops over each pixel 
        for pixel in image:
            
            if pixel_index < 754: # for each row except the last
                
                # takes average of the surrounding pixels
                surrounding_pixels_value = np.average(image[pixel_index+1] + image[pixel_index-1] +
                                                      image[pixel_index+28] + image[pixel_index+27] +
                                                      image[pixel_index+29] +
                                                      image[pixel_index-28] + image[pixel_index-27] +
                                                      image[pixel_index-29])
            else: # for the last row
                
                # takes average of the surrounding pixels
                surrounding_pixels_value = np.average(image[pixel_index-1] +
                                                      image[pixel_index-28] + image[pixel_index-27] +
                                                      image[pixel_index-29])
            
            # determines new pixel weight by linear conbination with coefficient weight
            new_pixel = weight*pixel + (1-weight)*surrounding_pixels_value
            
            # replaces old pixel with the new pixel value
            new_image[pixel_index] = new_pixel
            
            # increments pixel location in the image
            pixel_index += 1
            
        # append new image
        filtered_dev_data[image_index] = new_image
        
        # increments the counter
        image_index += 1
        
    # create instance of k nearest neighbors classifier
    classifier_three = KNeighborsClassifier(k)

    # generate model from mini training data
    classifier_three.fit(mini_train_data, mini_train_labels)

    # generate predictions for filtered dev images
    predictions_three = classifier_three.predict(filtered_dev_data)
        
    # generates report for the model
    report_three = sklearn.metrics.classification_report(dev_labels, predictions_three, output_dict = True)

    # overall accuracy for the model
    accuracy_three = np.round(100*report_three["accuracy"], 1)
    
    print("Model three (dev data filtered) accuracy: " + str(accuracy_three))
    
    
    
    ### Fourth Model ###
                
    # create instance of k nearest neighbors classifier
    classifier_four = KNeighborsClassifier(k)

    # generate model from filtered training data
    classifier_four.fit(filtered_train_data, mini_train_labels)

    # generate predictions for filtered dev images
    predictions_four = classifier_four.predict(filtered_dev_data)
        
    # generates report for the model
    report_four = sklearn.metrics.classification_report(dev_labels, predictions_four, output_dict = True)

    # overall accuracy for the model
    accuracy_four = np.round(100*report_four["accuracy"], 1)
    
    print("Model four (both filtered) accuracy: " + str(accuracy_four))


### STUDENT END ###

P6()
```

### Part 7:

Produce two Naive Bayes models and evaluate their performances.  Recall that Naive Bayes estimates P(feature|label), where each label is a categorical, not a real number.

For the first model, map pixel values to either 0 or 1, representing white or black - you should pre-process the data or use `BernoulliNB`'s `binarize` parameter to set the white/black separation threshold to 0.1.  Use `BernoulliNB` to produce the model.

For the second model, map pixel values to either 0, 1, or 2, representing white, gray, or black - you should pre-process the data, seting the white/gray/black separation thresholds to 0.1 and 0.9.  Use `MultinomialNB` to produce the model. 

Show the Bernoulli model accuracy and the Multinomial model accuracy.

Notes:
* Train on the mini train set.
* Evaluate performance on the dev set.
* `sklearn`'s Naive Bayes methods can handle real numbers, but for this exercise explicitly do the mapping to categoricals. 

Does the multinomial version improve the results? Why or why not?

```python
def P7():

### STUDENT START ###

    ### First Model ###

    # create classifier and binarize the images
    classifier = BernoulliNB(binarize = 0.1)
    
    # generate model from training data
    classifier.fit(mini_train_data, mini_train_labels)

    # generate predictions for dev images
    predictions = classifier.predict(dev_data)
        
    # generates report for the model
    report = sklearn.metrics.classification_report(dev_labels, predictions, output_dict = True)

    # overall accuracy for the model
    accuracy = np.round(100*report["accuracy"], 1)
    
    print("Bernoulli model accuracy: " + str(accuracy))
    
    
    ### Second Model ###
    
    # create empty array to hold mapped training data
    filtered_train_data = np.empty_like(mini_train_data)
    
    # create empty array to hold mapped dev data
    filtered_dev_data = np.empty_like(dev_data)
    
    # counter to keep track of image location in the dataset
    image_index = 0
    
    # repeats mapping process twice for training and dev data
    for dataset, new_dataset in [(mini_train_data, filtered_train_data), 
                                 (dev_data, filtered_dev_data)]:
    
        # loops over each image to apply the filter
        for image in dataset:

            # counter to keep track of pixel location in image
            pixel_index = 0

            # create blank new image
            new_image = np.empty_like(image)

            # loops over each pixel 
            for pixel in image:

                if pixel < 0.1: # first group

                    # map pixel to white
                    pixel = 0

                elif (pixel >= 0.1) and (pixel < 0.9): # second group

                    # map pixel to gray
                    pixel = 1

                else: # third group

                    # map pixel to black
                    pixel = 2

                # replaces old pixel with the new pixel value
                new_image[pixel_index] = pixel

                # increments pixel location in the image
                pixel_index += 1

            # append new image
            new_dataset[image_index] = new_image

            # increments the counter
            image_index += 1
          
        # reset image index
        image_index = 0
            
    # create classifier
    classifier_two = MultinomialNB()
    
    # generate model from training data
    classifier_two.fit(filtered_train_data, mini_train_labels)

    # generate predictions for dev images
    predictions_two = classifier_two.predict(filtered_dev_data)
        
    # generates report for the model
    report_two = sklearn.metrics.classification_report(dev_labels, predictions_two, output_dict = True)

    # overall accuracy for the model
    accuracy_two = np.round(100*report_two["accuracy"], 1)
    
    print("Multinomial model accuracy: " + str(accuracy_two))

    
### STUDENT END ###

P7()
```

ANSWER: The Multinomial model did not improve the results as it has a lower accuracy than the Bernoulli model. This result makes sense as the Bernoulli model explictly adds a penalty if it expects a certain feature of a class that is missing in a sample, while the multinomial model would ignore the lack of a feature.  

This is lack of improvement when further mapping to a third category also may have been because the threshold chosen did not add any meaningful features that the Bernoulli model wouldn't have already been able to use. Since most of each image has pixel intensity 0, corresponding to black, the added benefit of separating gray from white pixels likely just increased the noise the model would have to sort through.

Produce a Bernoulli Naive Bayes model and then use it to generate a 10x20 grid with 20 example images of each digit. Each pixel output should be either 0 or 1, based on comparing some randomly generated number to the estimated probability of the pixel being either 0 or 1.  Show the grid.

```python
def P10(num_examples):

### STUDENT START ###

    # create classifier and binarize the images
    classifier = BernoulliNB(binarize = 0.1)
    
    # generate model from training data
    classifier.fit(mini_train_data, mini_train_labels)
    
    # create square grid of images with num_examples examples
    # and 10 rows for each digit
    fig, axes = plt.subplots(10, num_examples, figsize = (10,10))
    
    # gets unique labels (i.e. digit names) from the data
    digits = np.unique(test_labels)
    
    # iterator for tracking how many digits have been completed
    current_digit = 0
    
    # probabilities for each label for each pixel
    model_prob = np.exp(classifier.feature_log_prob_)
    
    # iterate over each digit and its probability from the model
    for digit, probabilities in zip(digits, model_prob):

        # iterate for num_examples number of times
        for i in range(num_examples):
            
            # temp holding of the generated image
            generated_temp = np.empty_like(probabilities)

            pixel_probs = np.random.rand(784)
            
            # iterate over each pixel
            for j in range(len(pixel_probs)):
                
                # base case pixel is black
                pixel = 0
                
                # for case where probability is reached
                if pixel_probs[j] <= probabilities[j]:
                    
                    # makes pixel white
                    pixel = 1
                    
                # appends pixel to the generated image
                generated_temp[j] = pixel
    
              
            # reshapes data from 1D into a 28x28 image
            new_image = np.reshape(generated_temp, (28,28))
            
            # displays image on row with similar digits
            axes[current_digit, i].imshow(new_image)
            
            # sets colormap to black and white
            plt.rc("image", cmap = "gray")
            
            # turns off grid for subplot
            axes[current_digit, i].grid(False)
            
            # set y axis label for left most image
            if i == 0:
                axes[current_digit, i].set_ylabel(digit, fontsize = 20)
            
            # turns off x and y tick markers for subplot
            axes[current_digit, i].set_xticks([])
            axes[current_digit, i].set_yticks([])
            
        # increments the current digit
        current_digit += 1
            

### STUDENT END ###

P10(20)
```

### Part 11:

Recall that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior probability of the predicted class is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior probability and accuracy.  

Produce a Bernoulli Naive Bayes model.  Evaluate performance: partition the dev set into several buckets based on the posterior probabilities of the predicted classes - think of a bin in a histogram- and then estimate the accuracy for each bucket. So, for each prediction, find the bucket to which the maximum posterior probability belongs, and update "correct" and "total" counters accordingly.  Show the accuracy for each bucket.

Notes:
* Set LaPlace smoothing (alpha) to the optimal value (from part 8).
* Set binarization threshold to 0.
* Train on the mini train set.
* Evaluate perfromance on the dev set.

How would you characterize the calibration for this Bernoulli Naive Bayes model?

```python
def P11(buckets, correct, total):
    
### STUDENT START ###
    
    # create classifier
    classifier = BernoulliNB(alpha = nb.best_params_["alpha"])

    # generate model from training data
    classifier.fit(mini_train_data, mini_train_labels)

    # gets classes
    classes = classifier.classes_

    # iterates over each image in dev set
    for image, label in zip(dev_data, dev_labels):

        # prepares image for posterior probability function
        transformed_image = (np.array(image)).reshape(1, -1)
        
        # get posterior probability of the image for each class
        post_probs = classifier.predict_proba(transformed_image)[0]
        
        # get maximum posterior probability across all classes
        max_post_prob = np.amax(post_probs)
        
        # get digit associated with max posterior probability
        max_digit = np.argmax(post_probs)
        
        # placeholder for the index of the bucket the image
        # belongs to
        bucket_index = 0
        
        # iterate over each bucket with indexer
        for i in range(len(buckets)):
            
            # checks if post prob is still greater than current bucket
            if max_post_prob > buckets[i]:
                
                # continues searching for correct
                # post prob bucket
                continue
                
            # when post prob is less than the first bucket
            elif i == 0:
                
                # continues using zero for the bucket_index
                break
                
            # when post prob is less than a later bucket
            else:
                    
                # places image in previous bucket
                bucket_index = i - 1
                break
        
        # checks if prediction was correct
        if label == str(max_digit):
            
            # records correct prediction
            correct[bucket_index] += 1
            
        # increases total to the matching bucket
        total[bucket_index] += 1

              
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): 
        accuracy = correct[i] / total[i]
    print('p(pred) is %.13f to %.13f    total = %3d    accuracy = %.3f' % 
      (0 if i==0 else buckets[i-1], buckets[i], total[i], accuracy))
```

ANSWER: I would characterize this classifier as weakly calibrated. Throughout the bucket ranges, the accuracy increases with each increase in posterior probability. However, the accuracy does not track the actual value of the posterior probability well. For example, in the 90-99% range the accuracy only increases from 45.8% to 65.2%. While this is a large increase, it is still far away from the actual value of the posterior probability. The only time the accuracy and posterior probability align is in the first bucket between 0-50% posterior probability where the accuracy is 35.5%.
